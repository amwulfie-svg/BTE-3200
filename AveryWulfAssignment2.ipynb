```python
# Assignment 1- Quality Control

def getValues(n):
    values = []
    for i in range(n):
        val = float(input(f"Enter value #{i+1}: "))
        values.append(val)
    return values

def calcAverage(alist):
    if len(alist) <= 2:
        return 0
    usable_values = alist[2:]
    return sum(usable_values) / len(usable_values)

# Main Program
num = int(input("How many values will you enter? "))
weights = getValues(num)
avg = calcAverage(weights)
print("Average weight (excluding first two):", avg)
```

```python
# Assignment 3- Parking Garage

def getParkingHours():
    hours = float(input("Enter number of hours parked: "))
    return hours

def calcParkingFee(hours):
    fee = 2.50 * hours
    if fee < 6.00:
        fee = 6.00
    elif fee > 20.00:
        fee = 20.00
    return fee

# Main Program
hours = getParkingHours()
total_fee = calcParkingFee(hours)
print(f"Parking fee: ${total_fee:.2f}")
```

```python
# Assignment 4- Tuition

def calcTuition(credits):
    if credits >= 12:
        tuition = 20000
    else:
        tuition = 1200 + (credits * 1700)
    return tuition

# Main Program
credits = int(input("Enter number of credits: "))
tuition = calcTuition(credits)
print("Tuition for the semester: $", tuition)
```
